```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib
import cmocean.cm as cmo
```

```python
datadir = 'C:/Users/khana/Documents/Masters/Semester 2/Earth and System Modelling/PLASIM/PLASIM Data/'
era5_jan = xr.open_dataset(datadir + 'ERA5_jan.nc')
era5_jul = xr.open_dataset(datadir + 'ERA5_jul.nc')
pla_jan = xr.open_dataset(datadir + 'PLASIM_jan.nc')
pla_jul = xr.open_dataset(datadir + 'PLASIM_jul.nc')
```

```python
pe_jan = era5_jan.tp
pe_jul = era5_jul.tp
pm_jan = pla_jan.pr
pm_jul = pla_jul.pr

ue_jan = era5_jan.u10
ue_jul = era5_jul.u10
um_jan = pla_jan.ua
um_jul = pla_jul.ua

ve_jan = era5_jan.v10
ve_jul = era5_jul.v10
vm_jan = pla_jan.va
vm_jul = pla_jul.va

se_jan = era5_jan.si10
se_jul = era5_jul.si10
sm_jan = pla_jan.spd
sm_jul = pla_jul.spd
```

```python
#%% MONSOON HISTORICAL
plots = ['ERA5 January', 'ERA5 July', 'PlaSim January', 'PlaSim July']
ps = [pe_jan.sel(latitude=slice(55,-30),longitude=slice(230,330))*1000., pe_jul.sel(latitude=slice(55,-30),longitude=slice(230,330))*1000., pm_jan.sel(lat=slice(55,-30),lon=slice(230,330)), pm_jul.sel(lat=slice(55,-30),lon=slice(230,330))]
us = [ue_jan.sel(latitude=slice(45,-30),longitude=slice(230,330)), ue_jul.sel(latitude=slice(45,-30),longitude=slice(230,330)), um_jan.sel(lev=1000,lat=slice(45,-30),lon=slice(230,330)), um_jul.sel(lev=1000,lat=slice(45,-30),lon=slice(230,330))]
vs = [ve_jan.sel(latitude=slice(45,-30),longitude=slice(230,330)), ve_jul.sel(latitude=slice(45,-30),longitude=slice(230,330)), vm_jan.sel(lev=1000,lat=slice(45,-30),lon=slice(230,330)), vm_jul.sel(lev=1000,lat=slice(45,-30),lon=slice(230,330))]
ss = [se_jan.sel(latitude=slice(45,-30),longitude=slice(230,330)), se_jul.sel(latitude=slice(45,-30),longitude=slice(230,330)), sm_jan.sel(lat=slice(45,-30),lon=slice(230,330),lev=1000),sm_jul.sel(lat=slice(45,-30),lon=slice(230,330),lev=1000)]
```

```python
ssmax = 0.

for i in ss:
    val = i.max(skipna=True).values
    if val > ssmax:
        ssmax = val

vmin, vmax = 0, 20
fig, axs = plt.subplots(nrows=2,ncols=2,
                        subplot_kw={'projection': ccrs.PlateCarree()},
                        figsize=(17,12))
axs = axs.flatten()
for i, plot in enumerate(plots):
    data = ps[i]
    datau = us[i]
    datav = vs[i]
    datas = ss[i]
    if i < 2:
        lons, lats = np.meshgrid(data.longitude, data.latitude)
    else:
        lons, lats = np.meshgrid(data.lon, data.lat)
    pcm = axs[i].pcolormesh(lons, lats, data,
                            transform = ccrs.PlateCarree(),
                            cmap=cmo.rain, vmin=vmin, vmax=vmax)
    #Plot velocities
    if i < 2:
        lw = 2*datas.values / ssmax # set thickness streamlines
        axs[i].streamplot(datau.longitude, datau.latitude, datau.values, datav.values,
                          color='k', linewidth=10, transform=ccrs.PlateCarree())
    else:
        lw = 2*datas.values / ssmax
        axs[i].streamplot(datau.lon, datau.lat, datau.values, datav.values,
                          color='k', linewidth=10, transform=ccrs.PlateCarree())
    axs[i].set_title(plot)
    axs[i].coastlines()
    axs[i].set_extent([-120,-30,55,-30], crs=ccrs.PlateCarree())
fig.subplots_adjust(bottom=0.25, top=0.9, left=0.1, right=0.9,
                    wspace=0.02, hspace=0.1)
cbar_ax = fig.add_axes([0.2, 0.2, 0.6, 0.02])
cbar = fig.colorbar(pcm, cax=cbar_ax, orientation='horizontal')
cbar.set_label(label=r'total precipitation [mm/day]',fontsize=13)
plt.suptitle('Historical Mean Precipitation and Surface Winds (10m)',fontsize=18)
plt.show()
```

```python
#%% MONSOON DATA 1%/PI

# =============================================================================
# data = xr.open_dataset('atmo_monthly_postproc.nc') # PI T42 LSG
# =============================================================================

data = xr.open_dataset(datadir + 'onepct_monthly.nc') # PI T42 LSG
data = data.isel(time=slice(1104,None))

# Sum components
tpr = 24*60*60*1000*(data.prc.sel(lat=slice(55,-30),lon=slice(230,330))
                     + data.prl.sel(lat=slice(55,-30),lon=slice(230,330))
                     + data.prsn.sel(lat=slice(55,-30),lon=slice(230,330))) # total precipitation in mm/day
ua = data.ua.sel(lev=1000,lat=slice(55,-30),lon=slice(230,330))
va = data.va.sel(lev=1000,lat=slice(55,-30),lon=slice(230,330))
sa = (ua**2+va**2)**0.5

# January / July
month_idxs = tpr.groupby('time.month').groups
idxs = month_idxs[1]
tpr_jan = tpr.isel(time=idxs).mean(dim='time')
idxs = month_idxs[7]
tpr_jul = tpr.isel(time=idxs).mean(dim='time')

month_idxs = ua.groupby('time.month').groups
idxs = month_idxs[1]
ua_jan = ua.isel(time=idxs).mean(dim='time')
idxs = month_idxs[7]
ua_jul = ua.isel(time=idxs).mean(dim='time')

month_idxs = va.groupby('time.month').groups
idxs = month_idxs[1]
va_jan = va.isel(time=idxs).mean(dim='time')
idxs = month_idxs[7]
va_jul = va.isel(time=idxs).mean(dim='time')

month_idxs = sa.groupby('time.month').groups
idxs = month_idxs[1]
sa_jan = sa.isel(time=idxs).mean(dim='time')
idxs = month_idxs[7]
sa_jul = sa.isel(time=idxs).mean(dim='time')

#%% PLOTS PI/1pct DATA

plots = ['PlaSim January', 'PlaSim July']
ps = [tpr_jan, tpr_jul]
us = [ua_jan, ua_jul]
vs = [va_jan, va_jul]
ss = [sa_jan, sa_jul]

ssmax = 0.
for i in ss:
    val = i.max(skipna=True).values
    if val > ssmax:
        ssmax = val
#print(ssmax)

vmin, vmax = 0, 20

fig, axs = plt.subplots(nrows=1,ncols=2,
                        subplot_kw={'projection': ccrs.PlateCarree()},
                        figsize=(17,6))
axs = axs.flatten()
for i, plot in enumerate(plots):
    data = ps[i]
    datau = us[i]
    datav = vs[i]
    datas = ss[i]
    lons, lats = np.meshgrid(data.lon, data.lat)
    pcm = axs[i].pcolormesh(lons, lats, data,
                      cmap=cmo.rain, transform = ccrs.PlateCarree(), vmin=vmin, vmax=vmax)
    # Plot velocities
    lw = 2*datas.values / ssmax
    axs[i].streamplot(datau.lon, datau.lat, datau.values, datav.values,
                      color='k', linewidth=lw, transform=ccrs.PlateCarree())
    axs[i].set_title(plot)
    axs[i].coastlines()
    axs[i].set_extent([-120,-30,55,-30], crs=ccrs.PlateCarree())
fig.subplots_adjust(bottom=0.25, top=0.9, left=0.1, right=0.9,
                    wspace=0.02, hspace=0.1)
cbar_ax = fig.add_axes([0.2, 0.2, 0.6, 0.02])
cbar = fig.colorbar(pcm, cax=cbar_ax, orientation='horizontal')
cbar.set_label(label=r'total precipitation [mm/day]',fontsize=13)
#plt.suptitle('Pre-industrial Mean Precipitation and Surface Winds (10m)',fontsize=18)
plt.suptitle('1% Mean Precipitation and Surface Winds (10m)',fontsize=18)
plt.show()
```
